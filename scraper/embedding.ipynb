```python
import json

with open("../memes.json") as f:
    memes = json.load(f)

text_to_upsert = []
for meme in memes.values():
    text_to_upsert.append({
        "_id": meme["name"],
        "display_name": meme["display_name"],
        "description": meme["image_description"],
        "text_options":  [opt.get("description", "") for opt in meme["text_options"]],
    })

print(text_to_upsert)
```

```python
# Batch records in groups of 50
for i in range(0, len(text_to_upsert), 50):
    batch = text_to_upsert[i:i+50]
    index.upsert_records(
        "example-namespace",
        batch
    )
```

```python
results = index.search(
    namespace="example-namespace",
    query={
        "inputs": {"text": "the drake meme"}, 
        "top_k": 2
    },
    fields=["display_name", "description", "text_options"]
)

print(results)
```
